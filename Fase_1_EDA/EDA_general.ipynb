```python
# Mostrar los porcentajes nulos y el n√∫mero de valores nulos de cada columna
columnas = [
    'businesstravel', 
    'department', 
    'educationfield', 
    'hourlyrate', 
    'maritalstatus', 
    'monthlyincome', 
    'over18', 
    'overtime', 
    'performancerating', 
    'standardhours', 
    'totalworkingyears', 
    'worklifebalance', 
    'yearsincurrentrole', 
    'sameasmonthlyincome', 
    'salary', 
    'roledepartament', 
    'numberchildren'
]
```

```python
for col in columnas:
    procentajes_nulos = (df[col].isna().sum() / len(df)) * 100
    print(f'La columna {col.upper()} {df[col].isnull().sum()} tiene valores nulos lo que es igual a {procentajes_nulos.round()} %')
```
