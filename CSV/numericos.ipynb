```python
## Verificamos relacion entre columna salary y monthly income

df_numericos['is_equal'] = df_numericos.apply(lambda row: round(row['salary'] / 12,2) == round(row['monthlyincome'],2)
                          if pd.notnull(row['salary']) and pd.notnull(row['monthlyincome'])
                          else None, axis=1)
```
