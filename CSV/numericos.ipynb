```python
# Mover la coma una posición a la izquierda en las columnas 'dailyrate' y 'hourlyrate'
df_numericos['dailyrate'] = round(df_numericos['dailyrate'] / 10,2)
df_numericos['hourlyrate'] = round(df_numericos['hourlyrate'] / 10,2)
```

```python
## Tenemos que completar los 'NaN' de monthlyincome con la división de salary / 12 
# Solo actualizar las filas de 'monthlyincome' donde sea NaN
df_numericos['monthlyincome'] = df_numericos.apply(
    lambda row: row['salary'] / 12 if pd.notnull(row['salary']) and pd.isnull(row['monthlyincome']) else row['monthlyincome'],
    axis=1
)
```

```python
## Tenemos que completar los 'NaN' de salary con la multiplicacion de monthlyincome * 12 
# Solo actualizar las filas de 'salary' donde sea NaN
df_numericos['salary'] = df_numericos.apply(
    lambda row: row['monthlyincome'] * 12 if pd.notnull(row['monthlyincome']) and pd.isnull(row['salary']) else row['salary'],
    axis=1
)
```

```python
df_numericos['monthlyincome'] = round(df_numericos['monthlyincome'],2) ##redondeamos monthly income a dos decimales. 
```
