```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("unidos_limpio_1.csv", index_col=0)
```

```python
df.head()
```

```python
df.isna().sum()[df.isna().sum() > 0]
```

```python
#Comprobar el porcentaje de nulos.
lista_columnas = ["businesstravel", "department", "educationfield", "overtime", "performancerating", "worklifebalance"  ]

for col in lista_columnas:
    procentajes_nulos = (df[col].isna().sum() / len(df)) * 100
    print(f'La columna {col.upper()} {df[col].isnull().sum()} valores nulos lo que es igual a {procentajes_nulos.round()} %')
```

```python
df['educationfield'] = df['educationfield'].fillna('Unknown')
```

```python
df["educationfield"].isna().sum()
```

```python
valor_moda = df["overtime"].mode()[0]

df["overtime"].fillna(valor_moda)
```

```python
df.head(10)
```

```python
df["overtime"].isna().sum
```

```python
df.head()
```

```python
mediana = df["performancerating"].median()
df["performancerating"] = df["performancerating"].fillna(mediana)
```

```python
df["performancerating"] = df["performancerating"].astype(int)
```

```python
df["worklifebalance"].describe()[["mean", "50%"]]
```

```python
mediana = df["worklifebalance"].median()
df["worklifebalance"] = df["worklifebalance"].fillna(mediana)
```

```python
df["worklifebalance"] = df["worklifebalance"].astype(int)
```

```python
df["worklifebalance"].value_counts()
```

```python
df['businesstravel'] = df['businesstravel'].fillna('non travel')
```

```python
df["businesstravel"].unique()
```

```python
df['department'] = df['department'].fillna('Unknown')
```

```python
df.head()
```

```python
lista_minus = ["attrition","department","educationfield","overtime"]

for columna in lista_minus: 
    df[columna] = df[columna].str.lower()
```
