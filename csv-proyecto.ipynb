```python
import pandas as pd

# def limpiar_transformar_datos(df):
    # Reemplazar los valores de la columna 'Gender' de 0 y 1 a 'Male' y 'Female'
    if 'Gender' in df.columns:
        df['Gender'] = df['Gender'].map({0: 'Male', 1: 'Female'})

    # Convertir las columnas numÃ©ricas de tipo string a tipo numÃ©rico
    for column in df.select_dtypes(include=['object']).columns:  # Buscar columnas tipo string
        try:
            df[column] = pd.to_numeric(df[column], errors='coerce')  # Convertir a numÃ©rico
        except ValueError:
            pass  # Si no puede convertirse, pasa a la siguiente columna

    # Eliminar filas duplicadas
    df = df.drop_duplicates()

    # Corregir valores negativos en la columna 'DistanceFromHome' (reemplazar con NaN)
    if 'DistanceFromHome' in df.columns:
        df['DistanceFromHome'] = df['DistanceFromHome'].apply(lambda x: x if x >= 0 else None)

    # Corregir errores tipogrÃ¡ficos en la columna 'MaritalStatus' (reemplazar 'Marreid' por 'Married')
    if 'MaritalStatus' in df.columns:
        df['MaritalStatus'] = df['MaritalStatus'].replace({'Marreid': 'Married'})

    # Eliminar columnas redundantes (por ejemplo, columnas con el mismo nombre o informaciÃ³n)
    df = df.loc[:, ~df.columns.duplicated()]  # Eliminar columnas con nombres duplicados

    return df


# Aplicar la funciÃ³n
df_limpio = limpiar_transformar_datos(df)

# Mostrar el DataFrame limpio
print(df_limpio)
```
